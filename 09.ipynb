```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
#Plot Box Plot of Age Distribution by Gender and Survival
plt.figure(figsize=(10, 6))
sns.boxplot(data=titanic, x='sex', y='age', hue='survived')
plt.title("Age Distribution by Gender and Survival Status")
plt.xlabel("Sex")
plt.ylabel("Age")
plt.legend(title="Survived", labels=["No", "Yes"])
plt.grid(True)
plt.show()
```

```python
# Females had higher survival rates than males (the "Yes" boxes are more prominent for females).

# Median age of survivors is generally lower than that of non-survivors, especially among males.

# There is a wide age range among non-survivors, particularly in males (young children to elderly).

# Female survivors are often younger, suggesting lifeboats may have prioritized "women and children."
```
